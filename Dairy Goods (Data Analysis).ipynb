```python
import pandas as pd

df = pd.read_csv("dairy_dataset.csv")
df.head()
```

```python
df = df.drop(["Location","Quantity (liters/kg)","Price per Unit","Quantity Sold (liters/kg)","Price per Unit (sold)","Customer Location","Sales Channel","Quantity in Stock (liters/kg)","Minimum Stock Threshold (liters/kg)","Reorder Quantity (liters/kg)"], axis = 1)
df.head()
```

```python
df.info()
```

```python
#1. Does no. of cows in a farm has any dependency on area required.

a = df.groupby("Farm Size").agg({"Number of Cows" : "sum" , "Total Land Area (acres)" : "sum"})
a

x = a.corr()
x
```

```python
#2. What is the average expiry days of each product.

a = df.groupby("Product Name").agg({"Shelf Life (days)" : "mean"}) 
a
```

```python
#3. Mention the name of products with max & min shelflife

a = df.groupby("Product Name").agg({"Shelf Life (days)" : "min"})
a
# min shellife = milk (1 day)
```

```python
b = df.groupby("Product Name").agg({"Shelf Life (days)" : "max"})
b
# max shelf life = ghee (150 days)
```

```python
x = pd.merge(left = a, right = b, on = "Product Name", how = "inner")
x
```

```python
#4. Sumarize the data on the basis of brands and land acquired by them,cows & revenue generated by them.

df["Number of Cows"] = df["Number of Cows"].astype('float')

# which brand has generated max and min revenue with total no. of cows available
a = df.groupby("Brand").agg({"Total Land Area (acres)" : "sum","Number of Cows" : "sum","Approx. Total Revenue(INR)" : "sum"})
a
```

```python
#5. What is the average shelf life for different storage conditions.

a = df.groupby("Storage Condition").agg({"Shelf Life (days)" : "mean"})
a
```
